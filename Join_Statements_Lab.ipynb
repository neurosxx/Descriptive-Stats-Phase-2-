Hint: join the employees and offices tables. Select the first and last name.

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by. Select the office code, city, and number of employees.

```python
q = '''
SELECT o.officeCode, o.city, COUNT(e.employeeNumber) AS n_employees
FROM offices AS o
LEFT JOIN employees AS e USING(officeCode)
GROUP By officeCode
HAVING n_employees = 0;
'''
pd.read_sql(q,conn)

# Need to figure out how to reference officeCode as an alias for SQL to recognize which table you're referring to cus they both have office code column!
```

```python
# Note that COUNT(*) is not appropriate here because
# we are trying to count the _employees_ in each group.
# So instead we count by some attribute of an employee
# record. The primary key (employeeNumber) is a
# conventional way to do this
q = """
SELECT
    o.officeCode,
    o.city,
    COUNT(e.employeeNumber) AS n_employees
FROM offices AS o
LEFT JOIN employees AS e ON e.officeCode = o.officeCode
GROUP BY o.officeCode
HAVING n_employees = 0
;
"""
pd.read_sql(q, conn)
# Referred on line 11 to make offices as o. Able to break python rule of declaring a variable before using it. As long as your variables are defined somewhere in the query, you're good.
```

## Level Up 1: Display the names of every individual product that each employee has sold

Hint: You will need to use multiple `JOIN` clauses to connect all the way from employee names to product names.

<img src='https://curriculum-content.s3.amazonaws.com/data-science/images/Database-Schema.png' width="600">

```python
q = """
SELECT p.productName, e.lastName, e.firstName
FROM products AS p
JOIN orderdetails AS d ON p.productcode=d.productCode
JOIN orders AS o ON d.orderNumber=o.orderNumber
JOIN customers AS c ON o.customerNumber=c.customerNumber
JOIN employees AS e ON c.salesRepEmployeeNumber=e.employeeNumber
;
"""
pd.read_sql(q, conn)

# Look at CRM ERD image and follow the graph from employee names to product names (NOTICE how this has "all the way" in the instructions)
# Have to use JOIN multiple times, dont' make this too hard.
# Step 1: select columns you want to display, Notice how you define products AS p in 3rd line, able to reference on line 2 wiht p.productname
# Step 2: Follow table towards employees table, starting with joining orderdetails table on products with common key.
# Step 3 keep doing step 2, looks confusing but it's actuall linear. We then join orders on orderdetails. kl
```

## Level Up 2: Display the number of products each employee has sold

Alphabetize the results by employee last name.

Hint: Use the `quantityOrdered` column from `orderDetails`. Also, think about how to group the data when some employees might have the same first or last name.

```python
q = """
SELECT SUM(quantityOrdered) AS total_products_sold, e.lastName, e.firstName
FROM products AS p
JOIN orderdetails AS d ON p.productcode=d.productCode
JOIN orders AS o ON d.orderNumber=o.orderNumber
JOIN customers AS c ON o.customerNumber=c.customerNumber
JOIN employees AS e ON c.salesRepEmployeeNumber=e.employeeNumber
GROUP By e.employeeNumber
ORDER by e.lastName
;
"""
pd.read_sql(q, conn)

# Ask about shorter way to do it from Solution.  Also, ask why Order by and group by are benefiical her.e
```

## Level Up 3: Display the names of employees who have sold more than 200


different products

Hint: this is different from the previous question because the quantity sold doesn't matter, only the number of different products

```python
q = """
SELECT COUNT(d.productCode) AS different_products, e.lastName, e.firstName
FROM orderdetails AS d
JOIN orders AS o USING(orderNumber)
JOIN customers AS c USING(customerNumber)
JOIN employees AS e ON c.salesRepEmployeeNumber=e.employeeNumber
GROUP By e.lastName
HAVING different_products > 200
ORDER by e.lastName
;
"""

pd.read_sql(q, conn)

# LIne 6 needs ON statement due to not exact same key. Able to use "USING" When same key!
# Ask about shorter way to do it from Solution.  Also, ask why Order by and group by are benefiical here.
```
